```python
# If 1000 agents worth one million, have normally distributited geometric growth rates... 

n = 1000


mu0 = 6
x0 = mu0*np.ones(n)

t = []
for m in range(40000):
    x1 = np.random.normal(0, 0.2, n)
    t += [np.power(10, x0 + x1).sum()/(n*10**mu0)]
    
s1 = np.array(t)

t = []
for m in range(40000):
    x1 = np.random.normal(0, 0.1, n)
    t += [np.power(10, x0 + x1).sum()/(n*10**mu0)]
s2 = np.array(t)

plt.hist(np.log10(s1), 70, alpha = .5)
plt.hist(np.log10(s2), 70, alpha = .5)
plt.yscale('log')
plt.xlim(-0.15, 0.2)

plt.show()


n = 10

mu0 = 6
x0 = mu0*np.ones(n)

t = []
for m in range(40000):
    x1 = np.random.normal(0, 0.2, n)
    t += [np.power(10, x0 + x1).sum()/(n*10**mu0)]
    
s1 = np.array(t)

t = []
for m in range(40000):
    x1 = np.random.normal(0, 0.1, n)
    t += [np.power(10, x0 + x1).sum()/(n*10**mu0)]
s2 = np.array(t)

plt.hist(np.log10(s1), 70, alpha = .5)
plt.hist(np.log10(s2), 70, alpha = .5)
plt.yscale('log')
plt.xlim(-0.15, 0.2)

plt.show()
```

```python
def distrib(sigmas):
    result = np.log10(np.array([np.power(10, np.random.normal(0, s, 40000)) for s in sigmas]).sum(0)/len(sigmas))
    return result
```

```python
plt.hist(distrib([.1, .1]), 100, alpha = .5)
plt.hist(distrib([.1, .1, .1]), 100, alpha = .5)
plt.hist(distrib([.1, .1, .1, .1]), 100, alpha = .5)
plt.hist(distrib([.1, .1, .1, .1, .1, .1, .1, .1]), 100, alpha = .5)
# plt.xscale('log')
plt.yscale('log')
plt.show()
```

## Additive (usual) case

```python
mu0 = 3

s_i = 0.1
sigma = []
for n in range(1, 16):
    x0 = mu0*np.ones(n)

    t = []
    for m in range(10000):
        x1 = np.random.normal(0, s_i, n)
        t += [(x0 + x1).sum()/n]

    s2 = np.array(t)
    sigma += [np.std(s2)]
```

```python
y = np.array(sigma)
x = range(1, 16)

plt.plot(x, y, marker = '.')
plt.plot(x, s_i/np.sqrt(x), marker = '.')
plt.show()
```

```python
mu0 = 6

s_i = 0.18
sigma = []
mu = []
for i in range(5):
    n = 10**i
    
# for n in range(1, 16):
    
    x0 = mu0*np.ones(n)

    t = []
    for m in range(1000):
        x1 = np.random.normal(0, s_i, n)
#         t += [(x0 + x1).sum()/n]
        t += [np.power(10, x0 + x1).sum()/(n*10**mu0)]

    s2 = np.array(t)
    sigma += [np.std(s2)]
    mu += [np.mean(s2)]
```

```python
y = np.array(sigma)
# x = range(1, 16)
x = 10**np.arange(5)

plt.plot(x, y, marker = '.')
plt.plot(x, 2.4*s_i/np.sqrt(x), marker = '.')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
