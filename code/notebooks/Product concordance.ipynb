```python
import pandas as pd
```

```python
df = pd.read_csv('./../../data/processed/product_compare.csv')
```

```python
CN_full = pd.read_csv('./../../data/CN_full.csv', encoding = 'utf-8')
CN_4 = CN_full[['CN ID 4', 'CN label 4']].drop_duplicates()
CN_4.head()
```

```python
df = df.merge(CN_4, how = 'left')
```

```python
conc = df.groupby(['CPA6','CN ID 4','CN label 4','CN ID 8']).sum()
```

```python
from numpy import round

conc_pcts = conc.reset_index().groupby(['CN ID 4','CN label 4'])['VART'].apply(lambda x: 100 * x / float(x.sum()))
conc['pct'] = round(conc_pcts.values, 1)

pd.options.display.max_rows = 999
len(conc)

# conc.sort_index().head(1000)[['VART', 'pct']].style.bar(color='#d65f5f')
```

```python
# HS8 in CPA6

conc = df.groupby(['CN ID 8','CPA6','CN label 4']).sum()

conc_pcts = conc.reset_index().groupby(['CN ID 8'])['VART'].apply(lambda x: 100 * x / float(x.sum()))
conc['pct'] = round(conc_pcts.values, 1)

pd.options.display.max_rows = 999
len(conc)

conc.loc[conc.pct > 4].sort_index().head(1000)[['VART', 'pct']].style.bar(color='#d65f5f')
```
